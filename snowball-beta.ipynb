```python
tweet_words = " ".join(tweets['tweet_text'])
```

```python
for c in string.uppercase + string.lowercase: 
    STOPWORDS.add(c)
```

```python
# remove URLs, RTs, and twitter handles
no_urls_no_tags = [word for word in tweet_words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and word != 'RT' 
                                and word not in STOPWORDS
                            ]
```

```python
cloud = WordCloud().generate(" ".join(no_urls_no_tags))
```
